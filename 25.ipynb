```python
# for data analysis
import pandas as pd
#For Pre Processing Data
from sklearn import preprocessing
# To divide data in training and testing
from sklearn.model_selection import train_test_split
```

Importing Data set

```python
df=pd.read_csv('/content/drive/MyDrive/Assingnment01/malicious-website-dataset.csv')
```

Randomizing Data Set

```python
random_df=df.sample(frac=1)
random_df.reset_index(drop=True,inplace=False)
```

```python
random_df.describe()
```

Checking Nan and Null values in Data set

```python
random_df.isna().sum()
```

Filling Nan values with Mode

```python
random_df = random_df.fillna(random_df.mode())
```

Checking if still have Nan Values

```python
random_df.isna().sum()
```

Dropping Columns

```python
random_df.drop(columns=['CHARSET','SERVER','WHOIS_COUNTRY','WHOIS_STATEPRO','WHOIS_REGDATE','WHOIS_UPDATED_DATE'],inplace=True)
```

Checking Again

```python
random_df.isna().sum()
```

Dropping column URL

```python
random_df.drop(columns=['URL'],inplace=True)
```

Normalization

```python
X = random_df.iloc[:,:-1]
y = random_df.loc[:, ['Type']]
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(X)
normalize_df = pd.DataFrame(x_scaled)
```

```python
normalize_df
```

Spliting for Training

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
y_train
```

Training Classifier

```python
# train a Gaussian Naive Bayes classifier on the training set
from sklearn.naive_bayes import GaussianNB


# instantiate the model
gnb = GaussianNB()


# fit the model
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
y_pred_train = gnb.predict(X_train)

y_pred_train
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(gnb.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(gnb.score(X_test, y_test)))
```
